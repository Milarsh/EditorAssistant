```python
from sentence_transformers import SentenceTransformer
from transformers import pipeline
import numpy as np
```

```python
st = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
# bert = pipeline('feature-extraction', model='cointegrated/rubert-tiny2')
```

```python
def relevance(text, kwords):

    #sentence transformers
    st_text = st.encode(text)
    st_kwords = st.encode(kwords)

    def smlrty(a, b):
        return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

    scores = [smlrty(st_text, kword) for kword in st_kwords]

    best_idx = np.argmax(scores)

    return scores[best_idx] > 0.3, kwords[best_idx]
```

```python
rel, kw = relevance(
    "Tesla запустила новую автономную систему вождения",
    ["автомобили", "медицина", "искусственный интеллект", "стройка"]
)
print(f"Релевантно: {rel} - {kw}")
```
